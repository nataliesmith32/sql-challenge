```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
#create the db connection
engine = create_engine('postgres://postgres:postgres@localhost:5432/EmployeeSQL')
conn = engine.connect()
```

```python
#create salaries table
salaries_data = pd.read_sql("SELECT * FROM salaries", conn)
salaries_data.head()
```

```python
#create employee data table
employee_data = pd.read_sql("SELECT * FROM employee_data", conn)
employee_data.head()
```

```python
#combine employee and salary tables for histogram
combined_data = pd.merge(salaries_data, employee_data, on="emp_no", how="inner")
combined_data.head()
```

```python
#group by salary for histogram
emp_salary = combined_data.drop(columns = ["emp_title_id", "birth_date", "first_name", "last_name", "sex", "hire_date"])
emp_salary.head()
```

```python
#figuring out what our highest salary is
maxmoney = emp_salary.groupby(['salary']).max()
maxmoney.head(1)
```

```python
hist = plt.hist(emp_salary['salary'], bins=5)
plt.style.use('ggplot')
plt.xlabel("Average Salary Range")
plt.ylabel("Total Employees")

```

```python
#create title data table
title_info = pd.read_sql("SELECT * FROM titles", conn)
title_info.head()
```

```python
#rename column header to be able to merge tables
title_clean = title_info.rename(columns = {'title_id':'emp_title_id'}, inplace = False)
title_clean
```

```python
#combine employee and title tables for bar graph
combined_data1 = pd.merge(title_clean, employee_data, on="emp_title_id", how="inner")
combined_data1.head()
```

```python
title_group = combined_data1.groupby(['title']).mean()
title_group.rename(columns = {'emp_no':'avg_salary'}, inplace = True)
title_group.head()
```

```python
#set the index to call for graph
title_group = title_group.reset_index()
title_group
```

```python
#setting the axis and tick location
x_axis = title_group["title"]
ticks = np.arange(len(x_axis))
y_axis = title_group["avg_salary"]
 
#create the bar graph
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["k", "r", "g", "m", "b", "c", "y"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Position Titles")
plt.title("Average Salary by Employee Position")
```
